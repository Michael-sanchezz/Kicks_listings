```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
display(df)
```

```python
1. # Filter between SizeRank 1 and 10 (excluding "United States")
filtered_df = df[(df['SizeRank'] >= 1) & (df['SizeRank'] <= 10) & (df['RegionName'] != 'United States')]

print(filtered_df['RegionName'])
```

```python
# 2.Sort by home values in descending order.

rev_df =df.copy()
rev_df = rev_df.transpose()
# rev_df.head(7)
rev_df = rev_df.apply(lambda row: row.iloc[4:].sort_values(), axis=1)
display(rev_df)
# rev_df["avg"]= average_after_row_5 = df.iloc[2:].mean()
# print("Average of all columns after row 5:")
# rev_df.head(8)
```

```python
# 3.Format home values as currency with 2 decimal places.

columns_5_to_max = df.iloc[:, 4:] 
columns_5_to_max_rounded = columns_5_to_max.round(2)
df.iloc[:, 4:] = columns_5_to_max_rounded
display(df)
```

```python
# 4.Create a bar chart from the the data.

top_10_regions = df.sort_values(by='SizeRank').head(10)

# Create a bar chart
top_10_regions.plot(kind='bar', x='RegionName', y="2024-01-31", color='skyblue')
plt.grid(axis='y',color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Region Name')
plt.ylabel('Home Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# 5.Use MathPlotlib to reduce the gap between columns.

# Create a bar chart
top_10_regions.plot(kind='bar', x='RegionName', y='2023-04-30', color='skyblue',width = 0.8)
plt.grid(axis='y',color='gray', linestyle='--', linewidth=0.5)

plt.xlabel('Region Name')
plt.ylabel('Home Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6.Show Data Labels, include a Title, remove GridLines, and other formatting alterations to make an imagine similar to  the following image:

# Set the figure size
plt.figure(figsize=(16, 10))  # Adjust the width and height as needed

# Plot the bar chart with wider bars and more space between them
ax = top_10_regions.plot(kind='bar', x='RegionName', y='2000-01-31', color='orange', width=0.9, align='center',)
# Add prices at the top of each bar with smaller text
for p in ax.patches:
    ax.annotate('${:.2f}'.format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', xytext=(0, 5), 
                textcoords='offset points', fontsize=6)  # Adjust font size
# Add title and labels
plt.title('Home Values for the 10 Most Populous Regions')
plt.xlabel('Region Name')
plt.ylabel('Home Value')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
df_adi = pd.read_excel("C:/Users/micha/PycharmProjects/Docs/Adidas US Sales Datasets.xlsx") 
# Set the first row as column names
df_adi.columns = df_adi.iloc[3]
# Drop the first row
df_adi = df_adi.drop(3)
display(df_adi)
```

```python
df_adi.drop(df_adi.columns[0], axis=1,inplace=True)
df_adi.drop_duplicates(inplace=True)
display(df_adi)
```

```python

# Group by Price per Unit and calculate the sum of Units Sold
df_grouped = df_adi.groupby('Price per Unit')['Units Sold'].sum().reset_index()

# Plotting the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df_grouped['Price per Unit'], df_grouped['Units Sold'])

# Set labels and title

plt.ylabel('Sum of Units Sold')
plt.title('Price vs Units Sold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_grouped = df_adi.pivot_table(index='Product', columns='Region', values='Total Sales', aggfunc='sum')

print(df_grouped)
```

```python
df_grouped = df_adi.pivot_table(index='Product', columns='Region', values='Total Sales', aggfunc='sum')

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
df_grouped.plot(kind='bar', stacked=True)

# Set labels and title
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

plt.title('Top Selling Retailers by Region', fontsize=14)

# Show legend
plt.legend(title='Region', fontsize=10)

# Show grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.ticker as ticker
# Group by Product and Region, and sum the Total Sales
df_grouped = df_adi.pivot_table(index='Product', columns='Region', values='Total Sales', aggfunc='sum')

# Calculate total sales for each product
df_grouped['Total'] = df_grouped.sum(axis=1)

# Sort by total sales in descending order
df_grouped = df_grouped.sort_values(by='Total', ascending=False)

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
df_grouped.drop(columns='Total').plot(kind='bar', stacked=True)

# Set labels and title
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.title('Top Selling Retailers by Region', fontsize=14)

# Show legend
plt.legend(title='Region', fontsize=10)

# Show grid
plt.grid(True)

# Format y-axis as currency
formatter = ticker.FormatStrFormatter('$%1.0f')
plt.gca().yaxis.set_major_formatter(formatter)

# Show plot
plt.show()
```

```python
# Group by Sales Method and calculate the sum of Total Sales
df_grouped = df_adi.groupby('Sales Method')['Total Sales'].sum().reset_index()

# Format sales as currency
df_grouped['Total Sales'] = df_grouped['Total Sales'].map('${:,.0f}'.format)

# Plotting the bar chart
print(df_grouped)
```

```python

# Group by Sales Method and calculate the sum of Total Sales
df_grouped = df_adi.groupby('Sales Method')['Total Sales'].sum().reset_index()

# Plotting the bar chart with sideways bars
plt.figure(figsize=(10, 6))
bars = plt.barh(df_grouped['Sales Method'], df_grouped['Total Sales'], color='skyblue')

# Add data labels inside each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'${width}', ha='left', va='center', fontsize=10, color='black')

# Remove x-axis labels and ticks
plt.tick_params(axis='x', labelbottom=False)

# Set y-axis label and title
plt.ylabel('Sales Method', fontsize=12)
plt.title('Breakdown of Sales Methods', fontsize=14)

# Remove legend
plt.legend().remove()

# Show plot
plt.show()
```
